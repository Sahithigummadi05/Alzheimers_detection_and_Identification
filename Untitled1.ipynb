```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np
import os
import zipfile
from google.colab import drive

# 1. Mount Drive
drive.mount('/content/drive')

# 2. Check GPU
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))

# 3. Smart Unzip (Only runs if dataset folder is missing)
zip_path = "/content/drive/MyDrive/data.zip"
extract_to = "/content/dataset"

if not os.path.exists(extract_to):
    print(" Unzipping dataset...")
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)
    print(" Unzip complete!")
else:
    print(" Dataset already exists. Skipping unzip.")

# 4. Check contents (Important for next step!)
print("Folder contents:", os.listdir(extract_to))

DATA_DIR = "/content/dataset/Data"

print(f"Looking for images in: {DATA_DIR}")
print("Sub-folders found:", os.listdir(DATA_DIR))

IMG_SIZE = (128, 128)
BATCH_SIZE = 32

# 1. Load Training Data
train_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

# 2. Load Validation Data
val_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

class_names = train_ds.class_names
print("\n Classes found:", class_names)

#checking whether the pics are uploaded crctly or not
plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
plt.show()





```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# 1. Create Data Augmentation (Artificial Variety)
# This makes your dataset feel 10x larger
data_augmentation = models.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
])

# 2. Load the Pre-trained "Expert" (EfficientNetB0)
# include_top=False means we remove the last layer (which was for 1000 objects)
# weights='imagenet' loads the knowledge
base_model = tf.keras.applications.EfficientNetB0(
    include_top=False,
    weights="imagenet",
    input_shape=(128, 128, 3)
)

# Freeze the base model (We don't want to destroy its knowledge yet)
base_model.trainable = False

# 3. Combine everything
inputs = tf.keras.Input(shape=(128, 128, 3))
x = data_augmentation(inputs)  # Apply augmentation
x = base_model(x, training=False) # Pass through expert
x = layers.GlobalAveragePooling2D()(x) # Summarize features
x = layers.Dropout(0.2)(x)  # Regularization
outputs = layers.Dense(4, activation="softmax")(x) # 4 Classes

model = models.Model(inputs, outputs)

# Compile with a lower learning rate for stability
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=["accuracy"],
)

print(" Pro Model Built Successfully!")




from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau

# Define the "Managers"
callbacks = [
    # Save the BEST model to a file
    ModelCheckpoint(
        filepath="best_alzheimer_model.keras",
        save_best_only=True,
        monitor="val_accuracy"
    ),
    # Stop if not improving for 5 epochs
    EarlyStopping(
        monitor="val_accuracy",
        patience=5,
        restore_best_weights=True
    ),
    # Lower learning rate if stuck
    ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.2,
        patience=3
    )
]

# START TRAINING
print("Training Started... This may take a few minutes.")
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=20, # We set 20, but EarlyStopping might stop it sooner!
    callbacks=callbacks
)

import matplotlib.pyplot as plt

# Plot accuracy
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.legend()

plt.show()
```
